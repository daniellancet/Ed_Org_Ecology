```python
scholarships = df[df["Main_Category"] == 'Scholarship']
charters = df[df["Main_Category"] == 'Charter Schools']
k_12 = df[df["Main_Category"] == 'K-12 Schools']
fundraising = df[df["Main_Category"] == 'Fundraising']
PTG = df[df["Main_Category"] == 'Parent Teacher Groups']

dfs = [scholarships, charters, k_12, fundraising, PTG]
```

```python

from matplotlib_venn import venn3, venn3_circles

def describe_nulls(df, title):
    # Create boolean arrays for null values in each column
    tot_assets = df["total_assets"].isna()
    tot_income = df["total_income"].isna()
    totrev = df["totrev"].isna()

    # Count the number of null values in each set and their overlaps
    set1 = set(df[tot_assets].index)
    set2 = set(df[tot_income].index)
    set3 = set(df[totrev].index)
    
    # Calculate the overlaps
    only_tot_assets = len(set1 - set2 - set3)
    only_tot_income = len(set2 - set1 - set3)
    only_totrev = len(set3 - set1 - set2)
    tot_assets_tot_income = len(set1 & set2 - set3)
    tot_assets_totrev = len(set1 & set3 - set2)
    tot_income_totrev = len(set2 & set3 - set1)
    all_three = len(set1 & set2 & set3)

    # Create the Venn diagram with adjusted colors and transparency
    venn = venn3(subsets=(only_tot_assets, only_tot_income, tot_assets_tot_income,
                          only_totrev, tot_assets_totrev, tot_income_totrev, all_three),
                 set_labels=('Total Assets', 'Total Income', 'Total Revenue'),
                 set_colors=('blue', 'green', 'red'), alpha=0.5)
    
    # Draw circles to emphasize the borders
    venn_circles = venn3_circles(subsets=(only_tot_assets, only_tot_income, tot_assets_tot_income,
                                          only_totrev, tot_assets_totrev, tot_income_totrev, all_three),
                                 linestyle='dashed', linewidth=1, color='black')
    
    # Show the plot
    plt.title("Venn Diagram of Null Values: " + title)
    plt.show()


```

```python
names = ["scholarship", "charter", "k-12", "fundraising", "ptg"]
for i, df in enumerate(dfs):
    describe_nulls(df, names[i] + " " + " Out of " + str(len(df)))
    
    
```
